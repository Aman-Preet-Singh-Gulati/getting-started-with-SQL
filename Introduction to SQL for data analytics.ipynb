## What is SQL?

Sql stands for Structured Query language, whenever we used to hear **`"SQL"`** straightaway first thing that comes to our mind is - **designing the database** but SQL is not only about designing and creating a database but it can be a very important tool in the case of **drawing insights** from the database itself, SQL sometimes turns to be an underrated skill when it comes to make business based decisions but it is actually one of the important tool to make some reliable and efficient decisions.

**Some `basic` database operations that SQL can perform:**

1. Creating a database
2. Adding new data in the existing database
3. Updating the existence data/schema
4. Segmenting the data (Group By clause with aggregate function)
5. Ordering the data based on the requirements (Order By clause)

## Difference between database administrator and database analyst

Peeps are often confused when they hear a job role like database administrator or database analyst though both kind of technical person have to work with database only so the foundation may be same but the task and the role defined for them is different. In this section we will discuss the key difference between both of the role and task which they perform.

### 1. Database Administrator

As this article is centric towards SQL for data analytics so the below mentioned points can be considered as something which data analyst need **not to perform** because it is not his/her goal.

* The abbreviation of database administrator is **DBA** and he is the person who take the complete responsibility to handle, control and manipulate the data when I say manipulate the data so here I mean that individual can `update`, `delete` the data and also can `change the schema` based on the requiremetns and authorities given to them.


* He/she is also responsible to **keep the database upto date** i.e. to keep visiting the database design and updating it based on current scenarios.


* The individual who have the certain authority can **manage the database accessibility** which means he/she will handle the authority of accessibility to the database from other users and thus helps in maintaining **security**.


### 2. Database Analyst 

Now this is something to look for because here the roles and responsibilities that I'll discuss will be based on the role of database analyst.

* Where DBA handles the database design after that **data analysis** tasks are performed by the **database analyst** and his/her key responsibilities are to extract the key insights related to business and give the solution based on the historic data.


* When DBA is updating the data in the database on the usual basis then he is not a right person to be responsible to **filter out** the data because that is one of the key jobs in the profile of data analyst as he/she is responsible to extract all the data from the database and use effective **queries** to filter out only certain **chunks** of data according to certain requirements


* One of the most important role of data analyst is to **clean the data** as all the data which gets updated in the database is not the relevant data hence it becomes essential to **clean** and **transform** it for further analysis.

## How is SQL used in data engineering?

Before understanding the use of SQL in data engineering we first need to understand that what is the role of **Data Engineer**.
Key role of data engineers are:

* **Finding key insights**: One of the work for data enginner is to find the relavent insight from the whole database.
* **Interpreting data**: Sometimes the data which is extracted from the database is not understandable so cleaning and transforming the same is done by a data engineer.
* **ETL process**: Expanded form of ETL is **extract, load** and **transform** the data and in this process data engineer plays the key role.

The last point discussed as the role is where the SQL is used extinctively i.e. The whole process of ETL: Extracting, Loading and Transformation of data.

### 1. SELECT and FROM

This keyword forms one of the most basic SQL query and at the same time very important too because without these keywords doesn't matter how much analysis we have done and how many segments we have created the **record will not be shown**. Yes you guessed that right! this query will be responsible to **retrieve the data** from the database based on some conditions.


Note: In this article I'll be working on the database which have student's related data.
* Database name: "stm"
* Columns in the database:
    - employee
    - students
    
Now, let's see an example of SELECT and FROM statement:


SELECT * FROM employee;

where "employee" is the name of the table


Output:

Inference: So from the above output it is clear that we have **retrieved** the whole table by using * just after the SELECT statement. I know the question will be arise here that what if we want only selected fields from the table? 

It's actually very easy just replace the * with the name of the fields/field that you want to retrieve.

Syntax: 

SELECT

    field_1,
    field_2

FROM Table_name;

### 2. Distinct

Distinct statement will return the records from the **fields** which only have **unique values**.

Let's see an example:

SELECT distinct

	f_name

FROM employee;

Output:

Inference: If we compare the above output and the previous output so we can see that in previous output the **f_name** column was showing all the records i.e even the repeated values as well (**Aman**) but when we have used **DISTINCT** so it has removed all the **redundant values** and returned only the **unique values**.

### 3. Where clause

This is the ultimate clause which will help us to **filter the records** based on some **logical conditions** and it will return only those records which will satisfy those logical conditions.

In this particular example we will be working in the **student's table**. So first see that table's schema and what data it have.

SELECT * FROM students;

Output: 

Now let's see the example of where clause on the same table:


SELECT

	 name,
     email

FROM students

WHERE gender = 'Male';

Output:

Inference: From the above output we can see that "name" and "email" field is retrieved from the above query but main thing to notice is that the data that we can see is the **filtered** data which has been filtered with the help of **WHERE** clause based on the logical comdition which was that only **Male**'s data will be retrieved.

### 4. Order By clause

The Order By is query keyword which helps us to **sort** the records either in the form of ascending order or the descending order. Keeping the theoritical part away let's just simply see the example for the same.

SELECT *

FROM students

ORDER BY roll_no DESC;

Output:

Inference: In the above output we can clearly see that the data is sorted in the **descending** order by using the **ORDER BY** clause and **DESC** and if we want to sort back in the **ascending** order then the we will replace the **DISC** with **ASC** though we need to know one thing that the ascending order is set by default.

### 3. AS

**As** is the keyword which is formally known as **Aliases** in SQL and in mostly procedural oriented language. Aliases allows us to assign the custom name to a field.

Let's see an example:

SELECT 

	 name,
     email AS student_email

FROM students

WHERE gender = 'Male';

Output:

Inference: In the above output if we will see the field name of **email** it is now changed to **student_email** because we use the ALIAS (AS) to change the name of the column though it won't create any changes in the records this query is basically to make our SQL code more readable and understandable when we are using some **complex queries.**
